# histogram plot

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.random.randint(10, 40, (30))
y = np.random.randint(5, 30, (25))

#titles on the plot
plt.title ("Sample Graph", size = 15)
plt.xlabel ("random sample 2", size = 15)
plt.ylabel ("random sample 1", size = 15)

#calling histogram plot
plt.hist (x, edgecolor="b", color="y", range=(0,60), cumulative=-1, orientation="horizontal", label = "random")
plt.legend()
plt.show()
```

# Scatter plot

```python
x = [5,10,15,20,25,30]
y = [10,15,5,20,25,18]

#titles on the plot
plt.title("representation")
plt.xlabel("sample data")
plt.ylabel("number")

#calling scatter plot
plt.scatter(x,y)
plt.show()
```

# Pie Chart

```python
x = [10,20,30,40]
y = ["chips", "soft drink", "family pack", "sweets"]

#addons
ex = [0.0,0.0,0.0,0.08]
plt.title("Sample piechart")

#calling stem plot
plt.pie(x, labels=y, explode=ex, autopct="%0.1f%%", startangle=90, textprops={"size" : 13}, wedgeprops={"edgecolor": "Black"})
plt.show()
```

# Stem plot

```python
x = [1,2,4,5,6,3,2,5]
y = [0.5, 1, 2,3,4,5,6,7]

#titles
plt.title("sample representation")
plt.xlabel("random1")
plt.ylabel("random2")


#calling stem plot
plt.stem(x,y, linefmt= "-.", markerfmt= "ro", basefmt="purple", label="testdata")
plt.legend()
plt.show()
```

# Box plot

```python
import matplotlib.pyplot as plt

x = [5, 10, 15, 20, 25, 30, 35]
x1 = [10, 15, 20, 25]
x2 = [10.5, 15.5, 20.5, 25.5]
x3 = [10, 20, 25, 30]
y = [x, x1, x2, x3]

# Titles
plt.title("Box Representation", size=15)
plt.xlabel("Test Data", size=15)
plt.ylabel("Numbers", size=15)
lbl = ["Test1", "Test2", "Test3", "Test4"]

# Calling box plot 
boxplot = plt.boxplot(y, labels= lbl, boxprops=dict(color="red"), 
          capprops=dict(color="purple"), patch_artist=True)
           
# Change the colors as needed
colors = ['red', 'green', 'blue', 'orange']  
for patch, color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)

#displaying plot
plt.show()

```

# Stack plot

```python
x = [10,20,30,40,50]
y1 = [5,10,5,15,7]
y2 = [6,10,5,15,20]
y3 = [8,20,15,10,5]

#titles
plt.title("stack representation")
plt.xlabel("testdata")
plt.ylabel("testdata2")
lbl = ["y1", "y2", "y3"]

#calling plot
plt.stackplot(x,y1,y2,y3, labels=lbl, baseline="zero")
plt.legend(loc=2)
plt.show()
```

# Bar plot

```python
y = [10,5,20,15,20,30,10,35]
x = [5,10,15,20,25,30,35,40]
tick_label = ["A", "B","C","D","E","F","G","H"]


#titles
plt.title("Bar representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#calling bar-plot
plt.bar(x,y, tick_label=tick_label, color="Black",edgecolor="violet",linewidth=2, label="bars", alpha=0.5, hatch='///', width=5)
       
plt.legend()
plt.show()
```

# Step plot

```python
x = [2,4,6,8,10,12]
y = [2.5,4,6,8.5,10,11]

#titles
plt.title("Step representation")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#calling plot
plt.step(x,y, marker="o", color="red", label="test sample",mfc="aqua", ms=8)
plt.grid()
plt.legend()
plt.show()
```

# Fill between plot

```python
import numpy as np
x = np.array([2,4,6,8,10,12,14,16])
y = np.array([5,6,8,3,4,10,14,6])

#titles
plt.title("fill between plot representation")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#calling plot
plt.plot(x,y, label="sample_test")
plt.fill_between(x,y, color="violet", alpha=0.3, label="sample", where=(x>=4) & (x<=8))
plt.fill_between(x, y, color="violet", alpha=0.3, where=(x >= 10) & (x <= 16))
plt.legend()
plt.show()
```
